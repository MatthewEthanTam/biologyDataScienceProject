# 6.2 Importing the dataset
We will be focusing on tools used in data science to analyze the data. To do this, we will be using a python library called pandas. Pandas is a library that allows us to work with data in python. To import the dataset, we will use the following code:

Read through the comments in the next block to understand what the code is doing.


````python
# These are called comments and are not executed by the computer, for python comments start with a #
# This is an example on how to read in data from a csv file using pandas
# PS: a csv file is a comma separated value file, it is a common file format for data (kinda like excel)

# This imports the pandas library (https://pandas.pydata.org/docs/getting_started/index.html) and gives it an alias of pd, 
# you could choose any name you want, this allows us to use the pandas library by typing pd instead of pandas
import pandas as pd 

# Now we can use the read_csv function from pandas to read in the csv file
# The read_csv function takes in a file path to the csv file and returns a pandas dataframe

# Since we are in the lessons folder we need to go up 2 levels to the data folder and then into the ASD_microbiome.csv file
# The ../ means go up one level so 2 of them move up 2 level.

# The data is stored in a variable called ASD_data so we can use that variable to access the data
ASD_data = pd.read_csv("../../data/LS06-Data/ASD meta abundance.csv")

# Now we can use the head function to see the first 5 rows of the data
# The head function takes in a number of rows to display, if no number is given it defaults to 5
# the Display function is a special function that allows us to display the data in a nice format
display(ASD_data.head(5))
````

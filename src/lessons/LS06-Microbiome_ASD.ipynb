## 6.3.1 Calculating the mean, count, and standard deviation of the A3 column of the first dataset `ASD_data` (Method 1)

In this section we will be using python to calculate the mean, count, and standard deviation of the A3 column of the first dataset `ASD_data`.

Since the data is in a pandas dataframe it is much easier to do calculations on it. I will demonstrate two ways of getting the `mean`, `count`, and `standard deviation` of the A3 column of the first dataset `ASD_data`. 

There are many other ways and the first way is not the fastest or the most efficient but it will show you the logic of coding and then the second way is a more efficient way and mostly used by data scientists.



````python
# Calculating the mean of the A3 column
# Since the mean is calculated by adding all the values and dividing by the number of values
# We can do just that !

# I will be introducing a few things in this cell, 
# I suggest reading the following links to understand what each concept is:
#
# Concept of a `list` (https://www.w3schools.com/python/python_lists.asp)
# Concept of a `for loop` (https://www.w3schools.com/python/python_for_loops.asp)
# and the `len() function``  (https://www.w3schools.com/python/ref_func_len.asp)
# (NB: There are many ways to calculate the mean, this is just one of them
# that introduces some key concepts in programming)

# The first thing is to get all the values in the A3 column.

# To do this we can use the `.values` attribute of the ASD_data dataframe,
# and specify the column name we want to get the values from.

# This will return a list of all the values in the A3 column.

# Example, a list will look something like [1,2,3,4,5],

# each value is separated by a comma and the list is surrounded by square brackets.

column_A3 = ASD_data["A3"].values

# Now to count the amount of values in the list we can use the `len()` function.

count_of_values = len(column_A3)

# Now we can add all the values in the list together.

# To do this we declare a variable `sum_of_values` which will be used to store the sum of all the values.

# We set the value of the variable to 0 because the sum starts at 0 and we will add to it.

sum_of_values = 0

# To add each value in the list we can use a `for loop`.

# Below is a for loop that will go through each value in the column_A3 list and add it to the total variable.
# To explain the for loop in plain english: 

# There is a list of values called `column_A3` which look something like [1,2,3,4,5],
# We go through each value in the list and store it in the variable `i`.
# So for the example of [1,2,3,4,5], the first value of `i` would be 1, then 2, then 3, then 4, then 5.

# To break this down:
# The `for` keyword denotes the start of the for loop.
# The `i` variable is the variable that will be used to store the current value in the list.
# The `in` keyword is used to denote the list that will be used.
# The list is the `column_A3` list that we created earlier.
# The `:` denotes the end of the for loop declaration and the beginning of the area in which will be looped over.

# The next thing we see is that there is an indentation (or a tab), this is how python knows what code is in the for loop.
# and where the loops area ends is denoted by the end of the indentation and back to the level of the for loop.


for i in column_A3:
    # In Loop
    sum_of_values = sum_of_values + i

# Out of Loop

# To get the mean we will now divide the `total` by the `count_of_values`.
mean = sum_of_values/count_of_values

# Since we now have the mean, and count of values we can get the standard deviation.

# The standard deviation is calculated by taking the difference between each value and the mean,
# squaring it, adding all the squared differences together, dividing by the count of values,
# and then taking the square root of the result.

# To do this we will need to use the `sqrt` function from the `math` library.
# So we will need to import the math library, usually we would like to import all the libraries we need at the top of the file,
# but for this example we will import it here.

import math

# we must first initialize the variable `sum_of_squared_differences` to 0 and increase it each loop.
sum_of_squared_differences = 0

# Now to calculate the standard deviation we will use the same for loop as before but this time we will calculate the standard deviation.

for i in column_A3:
    # In Loop
    # Calculate the difference between the value and the mean
    difference = i - mean
    # Square the difference, the `**` denotes the power of something, so `2**2` is 2 squared
    squared_difference = difference ** 2
    # Add the squared difference to the total
    sum_of_squared_differences = sum_of_squared_differences + squared_difference

# Out of Loop

# Now we have the sum of the squared differences we can divide it by the count of values

# To do this we will need to convert the count of values to a float, this is because the count of values is an integer
# which means it can only be a whole number, but we need to divide by a decimal number.
# So we convert it to a float which is a decimal number.

count_of_values_float = float(count_of_values)

# Now we can divide the sum of squared differences by the count of values
# and then take the square root of the result

standard_deviation = math.sqrt(sum_of_squared_differences/count_of_values_float)

# Now we can show out the mean, count, and standard deviation.
# To show the values we will use the `print` function.
# The print function takes in a value and prints it out to the console.
# We would also like to add some text to the print out so we will use the `+` operator to add strings together.
# And we will use the `str()` function to convert the numbers to strings so we can add them to the text.

print("The mean is: " + str(mean))
print("The count of values is: " + str(count_of_values))
print("The standard deviation is: " + str(standard_deviation))
````

## 6.3.2 Calculating the mean, count, and standard deviation of the A3 column of the first dataset `ASD_data` (Method 2)

````python
# Instead of doing all the code ourselves, lets not reinvent the wheel 
# and use the built in functions from pandas to calculate the mean, count, and standard deviation.

# Thus we can use the `mean()`, `count()`, and `std()` functions from pandas to calculate the mean, count, and standard deviation.

# We must first specify the column we want to calculate the mean, count, and standard deviation of.
# We can do this by using the `[]` operator on the ASD_data dataframe and specifying the column name.

mean_pd = ASD_data["A3"].mean()
count_pd = ASD_data["A3"].count()
std_pd = ASD_data["A3"].std()

# Now we can print out the values

print("The mean from pandas is: " + str(mean_pd))
print("The count of values from pandas is: " + str(count_pd))
print("The standard deviation from pandas is: " + str(std_pd))
````
